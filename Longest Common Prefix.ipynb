## Problem Statement

Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string `""`.

## Examples

### Example 1
**Input:**  
`strs = ["flower", "flow", "flight"]`

**Output:**  
`"fl"`

---

### Example 2
**Input:**  
`strs = ["dog", "racecar", "car"]`

**Output:**  
`""`

**Explanation:**  
There is no common prefix among the input strings.

---

## Constraints

- `1 <= strs.length <= 200`
- `0 <= strs[i].length <= 200`
- `strs[i]` consists of only lowercase English letters if it is non-empty.

# Test Cases
---

```python
def test_longest_common_prefix(solution):
    test_cases = [
        (["flower", "flow", "flight"], "fl"),
        (["dog", "racecar", "car"], ""),
        (["interspecies", "interstellar", "interstate"], "inters"),
        (["a"], "a"),
        (["", ""], ""),
    ]

    for strs, expected in test_cases:
        result = solution.longestCommonPrefix(strs)
        assert result == expected, f"Failed for input {strs}: expected {expected}, got {result}"

    print("All test cases passed!")
```

# Approach
---

## Longest Common Prefix

### Strategy
- Assume the first string as the initial prefix
- Compare the prefix with each subsequent string
- Shrink the prefix until it matches the start of the current string

### Time Complexity
- **O(n Â· m)**
  - `n` is the number of strings
  - `m` is the length of the shortest string

### Space Complexity
- **O(1)**

```python
class Solution:
    def longestCommonPrefix(self, strs):
        # If the list is empty, return empty string
        if not strs:
            return ""

        prefix = ""

        # Take characters from the first string one by one
        for i in range(len(strs[0])):
            current_char = strs[0][i]

            # Compare this character with the same position in all strings
            for s in strs:
                # If index goes out of range or characters don't match
                if i >= len(s) or s[i] != current_char:
                    return prefix

            # If all strings matched, add character to prefix
            prefix += current_char

        return prefix
```

```python
solution = Solution()
test_longest_common_prefix(solution)
```
