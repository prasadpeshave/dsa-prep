# Plus One

## Problem Statement

You are given a large integer represented as an integer array `digits`, where each `digits[i]` is the ith digit of the integer. The digits are ordered from most significant to least significant in left-to-right order. The large integer does not contain any leading 0's.

Increment the large integer by one and return the resulting array of digits.

## Examples

### Example 1
**Input:**  
`digits = [1,2,3]`

**Output:**  
`[1,2,4]`

**Explanation:**  
The array represents the integer `123`. Incrementing by one gives `124`.

---

### Example 2
**Input:**  
`digits = [4,3,2,1]`

**Output:**  
`[4,3,2,2]`

**Explanation:**  
The array represents the integer `4321`. Incrementing by one gives `4322`.

---

### Example 3
**Input:**  
`digits = [9]`

**Output:**  
`[1,0]`

**Explanation:**  
The array represents the integer `9`. Incrementing by one gives `10`.

---

## Constraints

- `1 <= digits.length <= 100`
- `0 <= digits[i] <= 9`
- `digits` does not contain any leading 0's.

# Test Cases
---

```python
def test_plus_one(solution):
    test_cases = [
        ([1, 2, 3], [1, 2, 4]),
        ([4, 3, 2, 1], [4, 3, 2, 2]),
        ([9], [1, 0]),
        ([9, 9, 9], [1, 0, 0, 0]),
        ([0], [1]),
    ]

    for digits, expected in test_cases:
        result = solution.plusOne(digits)
        assert result == expected, f"Failed for input {digits}: expected {expected}, got {result}"

    print("All test cases passed!")
```

# Approach
---

## Increment Digit Array

### Strategy
- Convert the digit array into a string and then into an integer
- Increment the integer by one
- Convert the result back into a list of digits

### Time Complexity
- **O(n)**  
  - `n` is the number of digits

### Space Complexity
- **O(n)**  
  - Due to string and integer conversions


```python
class Solution:
    def plusOne(self, digits):
        sol = list(str(int("".join(str(ele) for ele in digits))+1))
        return [int(ele) for ele in sol]
```

```python
solution = Solution()
test_plus_one(solution)
```
