# Power of Two

**Problem Link:**  
https://leetcode.com/problems/power-of-two/

## Problem Statement
Given an integer `n`, return `true` if it is a power of two. Otherwise, return `false`.

An integer `n` is a power of two if there exists an integer `x` such that:
`n == 2^x`

## Examples

### Example 1
Input: `n = 1`  
Output: `true`

### Example 2
Input: `n = 16`  
Output: `true`

### Example 3
Input: `n = 3`  
Output: `false`

## Constraints
- `-2^31 <= n <= 2^31 - 1`

**Follow-up:** Solve without loops or recursion.


## Bit Manipulation (Optimal)

### Key Insight
A power of two has **exactly one set bit** in its binary representation.

Example:
- `1  -> 0001`
- `2  -> 0010`
- `4  -> 0100`
- `8  -> 1000`

For any power of two:


```python
class Solution:
    def isPowerOfTwo(self, n: int) -> bool:
        return n > 0 and (n & (n - 1)) == 0
```

```python
def test_power_of_two():
    sol = Solution()
    assert sol.isPowerOfTwo(1) == True
    assert sol.isPowerOfTwo(16) == True
    assert sol.isPowerOfTwo(3) == False
    assert sol.isPowerOfTwo(0) == False
    assert sol.isPowerOfTwo(-8) == False
    print("All test cases passed!")

test_power_of_two()
```
