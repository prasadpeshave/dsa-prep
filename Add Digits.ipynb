# Add Digits

**Problem Link:**  
https://leetcode.com/problems/add-digits/

---

## Problem Statement

Given an integer `num`, repeatedly add all its digits until the result has only **one digit**, and return it.

## Examples

### Example 1
**Input:** `num = 38`

**Output:** `2`

**Explanation:**  
`38 → 3 + 8 = 11 → 1 + 1 = 2`

---

### Example 2
**Input:** `num = 0`

**Output:** `0`

---

## Constraints

- `0 <= num <= 2³¹ - 1`

**Follow-up:** Solve without loops or recursion in `O(1)` time.

# Approach
---

## Digital Root (Math Trick)

### Key Insight
This problem is a classic **Digital Root** problem.

For any non-negative integer:
- If `num == 0`, result is `0`
- Otherwise, result is:

```
1 + (num - 1) % 9
```

### Why This Works
- Repeated digit sums eventually cycle every 9
- This formula directly computes the final single digit

### Time Complexity
- **O(1)**

### Space Complexity
- **O(1)**

````
class Solution:
    def addDigits(self, num: int) -> int:
        if num == 0:
            return 0
        return 1 + (num - 1) % 9
````

# Test Cases

````
def test_add_digits():
    sol = Solution()
    assert sol.addDigits(38) == 2
    assert sol.addDigits(0) == 0
    assert sol.addDigits(9) == 9
    assert sol.addDigits(18) == 9
    assert sol.addDigits(123) == 6
    print("All test cases passed!")

test_add_digits()
````
