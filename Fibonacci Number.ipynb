## Problem Statement

The Fibonacci numbers form a sequence such that each number is the sum of the two preceding ones, starting from `0` and `1`.

```
F(0) = 0
F(1) = 1
F(n) = F(n - 1) + F(n - 2), for n > 1
```

Given an integer `n`, calculate `F(n)`.

## Examples

### Example 1
**Input:**  
`n = 2`

**Output:**  
`1`

---

### Example 2
**Input:**  
`n = 3`

**Output:**  
`2`

---

### Example 3
**Input:**  
`n = 4`

**Output:**  
`3`

---

## Constraints

- `0 <= n <= 30`

# Test Cases
---

````python
def test_fib(solution):
    test_cases = [
        (0, 0),
        (1, 1),
        (2, 1),
        (3, 2),
        (4, 3),
        (10, 55),
    ]

    for n, expected in test_cases:
        result = solution.fib(n)
        assert result == expected, f"Failed for n={n}: expected {expected}, got {result}"

    print("All test cases passed!")
````

# Naive Recursive Approach
---

## Plain Recursion

### Strategy
- Directly follow the Fibonacci definition
- Recursively compute `F(n-1)` and `F(n-2)`

### Time Complexity
- **O(2â¿)**

### Space Complexity
- **O(n)** due to recursion stack

````python
class RecursiveSolution:
    def fib(self, n: int) -> int:
        if n <= 1:
            return n
        return self.fib(n - 1) + self.fib(n - 2)
````

````python
solution = RecursiveSolution()
test_fib(solution)
````

# Optimized Approach
---

## Iterative Dynamic Programming

### Strategy
- Build the Fibonacci sequence iteratively
- Store only the previous two values

### Time Complexity
- **O(n)**

### Space Complexity
- **O(1)**

````python
class OptimizedSolution:
    def fib(self, n: int) -> int:
        if n <= 1:
            return n

        prev, curr = 0, 1
        for _ in range(2, n + 1):
            prev, curr = curr, prev + curr

        return curr
````

````python
solution = OptimizedSolution()
test_fib(solution)
````
