## Problem Statement

You are given two integer arrays `nums1` and `nums2`, sorted in non-decreasing order, and two integers `m` and `n`, representing the number of valid elements in `nums1` and `nums2` respectively.

Merge `nums1` and `nums2` into a single array sorted in non-decreasing order.

**Important:** The result must be stored inside `nums1` in-place. `nums1` has a length of `m + n`, where the last `n` elements are placeholders (`0`).

## Examples

### Example 1
**Input:**  
`nums1 = [1,2,3,0,0,0], m = 3`  
`nums2 = [2,5,6], n = 3`

**Output:**  
`[1,2,2,3,5,6]`

---

### Example 2
**Input:**  
`nums1 = [1], m = 1`  
`nums2 = [], n = 0`

**Output:**  
`[1]`

---

### Example 3
**Input:**  
`nums1 = [0], m = 0`  
`nums2 = [1], n = 1`

**Output:**  
`[1]`

---

## Constraints

- `nums1.length == m + n`
- `nums2.length == n`
- `0 <= m, n <= 200`
- `-10⁹ <= nums1[i], nums2[i] <= 10⁹`
- Arrays are sorted in non-decreasing order

# Test Cases
---

```python
def test_merge(solution):
    test_cases = [
        ([1,2,3,0,0,0], 3, [2,5,6], 3, [1,2,2,3,5,6]),
        ([1], 1, [], 0, [1]),
        ([0], 0, [1], 1, [1]),
    ]

    for nums1, m, nums2, n, expected in test_cases:
        solution.merge(nums1, m, nums2, n)
        assert nums1 == expected, f"Failed: expected {expected}, got {nums1}"

    print("All test cases passed!")
```

# Approach
---

## Simple Merge Using Extra Space

### Strategy
- Extract the valid `m` elements from `nums1`
- Combine them with `nums2`
- Sort the merged list
- Copy the result back into `nums1`

### Time Complexity
- **O((m + n) log(m + n))** due to sorting

### Space Complexity
- **O(m + n)** due to extra list usage

```python
from typing import List

class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        merged = nums1[:m] + nums2
        merged.sort()
        nums1[:] = merged
```

```python
solution = Solution()
test_merge(solution)
```
