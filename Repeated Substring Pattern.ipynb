## Problem Statement

Given a string `s`, check if it can be constructed by taking a substring of it and appending multiple copies of the substring together.

## Examples

### Example 1
**Input:**  
`s = "abab"`

**Output:**  
`true`

**Explanation:**  
`"ab"` repeated twice forms `"abab"`.

---

### Example 2
**Input:**  
`s = "aba"`

**Output:**  
`false`

---

### Example 3
**Input:**  
`s = "abcabcabcabc"`

**Output:**  
`true`

**Explanation:**  
`"abc"` repeated four times forms the string.

---

## Constraints

- `1 <= s.length <= 10â´`
- `s` consists of lowercase English letters

# Test Cases
---

```python
def test_repeated_substring(solution):
    test_cases = [
        ("abab", True),
        ("aba", False),
        ("abcabcabcabc", True),
        ("a", False),
        ("zzzz", True),
    ]

    for s, expected in test_cases:
        result = solution.repeatedSubstringPattern(s)
        assert result == expected, f"Failed for {s}: expected {expected}, got {result}"

    print("All test cases passed!")
```

# Approach
---

## String Doubling Trick

### Strategy
- Create a new string by concatenating the original string with itself
- Remove the first and last characters from this new string
- If the original string exists inside this modified string, it is a repeated pattern

### Why This Works
- Repeated patterns reappear when the string is doubled
- Non-repeating strings will not reappear in the trimmed version

### Time Complexity
- **O(n)**

### Space Complexity
- **O(n)**

```python
class Solution:
    def repeatedSubstringPattern(self, s: str) -> bool:
        doubled = s + s
        return s in doubled[1:-1]
```

```python
solution = Solution()
test_repeated_substring(solution)
```
