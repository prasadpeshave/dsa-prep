# First Bad Version

**Problem Link:**

## Problem Statement

You are given `n` versions `[1, 2, ..., n]`. Once a version is bad, all following versions are also bad.

You are given an API `isBadVersion(version)` which returns whether a version is bad.

Your task is to find the **first bad version** while minimizing the number of API calls.

## Examples

### Example 1
**Input:**  
`n = 5, bad = 4`

**Output:**  
`4`

---

### Example 2
**Input:**  
`n = 1, bad = 1`

**Output:**  
`1`

---

## Constraints

- `1 <= bad <= n <= 2³¹ - 1`

# Mock API for Testing
---

```python
# Mock API for local testing
BAD_VERSION = 4

def isBadVersion(version: int) -> bool:
    return version >= BAD_VERSION
```

# Approach
---

## Binary Search

### Strategy
- Perform binary search on range `[1, n]`
- If `mid` is bad, move left
- If `mid` is good, move right

### Time Complexity
- **O(log n)**

### Space Complexity
- **O(1)**

```python
class Solution:
    def firstBadVersion(self, n: int) -> int:
        left, right = 1, n

        while left < right:
            mid = (left + right) // 2

            if isBadVersion(mid):
                right = mid
            else:
                left = mid + 1

        return left
```

```python
solution = Solution()
print(solution.firstBadVersion(5))
```
