## Problem Statement

Given an integer array `nums`, return `true` if any value appears at least twice in the array, and return `false` if every element is distinct.

## Examples

### Example 1
**Input:**  
`nums = [1,2,3,1]`

**Output:**  
`true`

---

### Example 2
**Input:**  
`nums = [1,2,3,4]`

**Output:**  
`false`

---

### Example 3
**Input:**  
`nums = [1,1,1,3,3,4,3,2,4,2]`

**Output:**  
`true`

---

## Constraints

- `1 <= nums.length <= 10⁵`
- `-10⁹ <= nums[i] <= 10⁹`

# Test Cases
---

```python
def test_contains_duplicate(solution):
    test_cases = [
        ([1,2,3,1], True),
        ([1,2,3,4], False),
        ([1,1,1,3,3,4,3,2,4,2], True),
        ([1], False),
    ]

    for nums, expected in test_cases:
        result = solution.containsDuplicate(nums)
        assert result == expected, f"Failed for input {nums}: expected {expected}, got {result}"

    print("All test cases passed!")
```

# Approach
---

## Using a Set

### Strategy
- Convert the list to a set
- Compare the length of the set with the length of the list
- If lengths differ, duplicates exist

### Time Complexity
- **O(n)**

### Space Complexity
- **O(n)** due to the set

```python
from typing import List

class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        return len(set(nums)) != len(nums)
```

```python
solution = Solution()
test_contains_duplicate(solution)
```
