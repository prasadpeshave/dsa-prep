# Remove Duplicates from Sorted List

**Problem Link:**  
https://leetcode.com/problems/remove-duplicates-from-sorted-list/

---

## Problem Statement

Given the `head` of a sorted linked list, delete all duplicates such that each element appears only once. Return the linked list sorted as well.

## Examples

### Example 1
**Input:**  
`head = [1,1,2]`

**Output:**  
`[1,2]`

---

### Example 2
**Input:**  
`head = [1,1,2,3,3]`

**Output:**  
`[1,2,3]`

---

## Constraints

- The number of nodes in the list is in the range `[0, 300]`
- `-100 <= Node.val <= 100`
- The list is guaranteed to be **sorted**

# Approach
---

## Using Extra Space (Set)

### Strategy
- Traverse the linked list node by node
- Use a set to keep track of values already seen
- Create a new linked list using a dummy head
- Add a node only if its value has not appeared before

### Time Complexity
- **O(n)**

### Space Complexity
- **O(n)** due to the additional set

```
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        seen = set()
        dummy = ListNode(0)
        current = dummy

        while head is not None:
            if head.val not in seen:
                seen.add(head.val)
                current.next = ListNode(head.val)
                current = current.next
            head = head.next

        return dummy.next
```
