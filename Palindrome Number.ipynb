## Problem Statement

Given an integer `x`, return `true` if `x` is a palindrome, and `false` otherwise.

## Examples

### Example 1
**Input:**  
`x = 121`

**Output:**  
`true`

**Explanation:**  
`121` reads the same from left to right and right to left.

---

### Example 2
**Input:**  
`x = -121`

**Output:**  
`false`

**Explanation:**  
From left to right it reads `-121`, but from right to left it becomes `121-`.

---

### Example 3
**Input:**  
`x = 10`

**Output:**  
`false`

**Explanation:**  
Reads `01` from right to left, which is not the same.

---

## Constraints

- `-2³¹ <= x <= 2³¹ - 1`
