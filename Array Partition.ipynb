## Problem Statement

Given an integer array `nums` of `2n` integers, group these integers into `n` pairs such that the sum of `min(ai, bi)` for all pairs is maximized. Return the maximized sum.

## Examples

### Example 1
**Input:**  
`nums = [1,4,3,2]`

**Output:**  
`4`

---

### Example 2
**Input:**  
`nums = [6,2,6,5,1,2]`

**Output:**  
`9`

---

## Constraints

- `nums.length == 2 * n`
- `1 <= n <= 10⁴`
- `-10⁴ <= nums[i] <= 10⁴`

# Corrected Approach
---

## Sort and Pair Adjacent Elements

### Strategy
- Sort the array in non-decreasing order
- Pair adjacent elements in the sorted array
- For each pair, the smaller element is always the first one
- Sum all elements at even indices

### Why This Works
- Sorting ensures that pairing close values maximizes the minimum of each pair
- Choosing the smaller element from each adjacent pair gives the maximum total sum

### Time Complexity
- **O(n log n)** due to sorting

### Space Complexity
- **O(1)** (excluding sort overhead)

```python
from typing import List

class Solution:
    def arrayPairSum(self, nums: List[int]) -> int:
        sorted_nums = sorted(nums)
        min_sum = 0

        for i in range(0, len(sorted_nums), 2):
            min_sum += sorted_nums[i]

        return min_sum
```

```python
solution = Solution()
test_array_pair_sum(solution)
```
