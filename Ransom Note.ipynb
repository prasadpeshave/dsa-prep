## Problem Statement

Given two strings `ransomNote` and `magazine`, return `true` if `ransomNote` can be constructed by using the letters from `magazine`, and `false` otherwise.

Each letter in `magazine` can only be used once in `ransomNote`.

## Examples

### Example 1
**Input:**  
`ransomNote = "a", magazine = "b"`

**Output:**  
`false`

---

### Example 2
**Input:**  
`ransomNote = "aa", magazine = "ab"`

**Output:**  
`false`

---

### Example 3
**Input:**  
`ransomNote = "aa", magazine = "aab"`

**Output:**  
`true`

---

## Constraints

- `1 <= ransomNote.length, magazine.length <= 10âµ`
- Strings contain only lowercase English letters

# Test Cases
---

```python
def test_can_construct(solution):
    test_cases = [
        ("a", "b", False),
        ("aa", "ab", False),
        ("aa", "aab", True),
        ("", "abc", True),
        ("abc", "abc", True),
    ]

    for ransomNote, magazine, expected in test_cases:
        result = solution.canConstruct(ransomNote, magazine)
        assert result == expected, (
            f"Failed for ransomNote={ransomNote}, magazine={magazine}: "
            f"expected {expected}, got {result}"
        )

    print("All test cases passed!")
```

# Approach
---

## Frequency Counting (Hash Map)

### Strategy
- Count the frequency of each character in `magazine`
- For each character in `ransomNote`, reduce its count
- If any required character is missing or exhausted, return `False`

### Why This Works
- Each character can only be used once
- Frequency comparison ensures correct usage

### Time Complexity
- **O(n + m)** where `n` is length of `ransomNote` and `m` is length of `magazine`

### Space Complexity
- **O(1)** (only 26 lowercase English letters)

```python
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        freq = {}

        for ch in magazine:
            freq[ch] = freq.get(ch, 0) + 1

        for ch in ransomNote:
            if freq.get(ch, 0) == 0:
                return False
            freq[ch] -= 1

        return True
```

```python
solution = Solution()
test_can_construct(solution)
```
