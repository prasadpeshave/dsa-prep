## Problem Statement

Given an integer `columnNumber`, return its corresponding column title as it appears in an Excel sheet.

Mapping:

- `A -> 1`
- `B -> 2`
- `...`
- `Z -> 26`
- `AA -> 27`
- `AB -> 28`

## Examples

### Example 1
**Input:**  
`columnNumber = 1`

**Output:**  
`"A"`

---

### Example 2
**Input:**  
`columnNumber = 28`

**Output:**  
`"AB"`

---

### Example 3
**Input:**  
`columnNumber = 701`

**Output:**  
`"ZY"`

---

## Constraints

- `1 <= columnNumber <= 2³¹ - 1`

# Test Cases
---

```python
def test_convert_to_title(solution):
    test_cases = [
        (1, "A"),
        (26, "Z"),
        (27, "AA"),
        (28, "AB"),
        (52, "AZ"),
        (701, "ZY"),
    ]

    for columnNumber, expected in test_cases:
        result = solution.convertToTitle(columnNumber)
        assert result == expected, (
            f"Failed for input {columnNumber}: expected {expected}, got {result}"
        )

    print("All test cases passed!")
```

# Approach
---

## Naive Approach – Base-26 Conversion

### Strategy
- Treat the column number as a base-26 value
- Subtract `1` at each step to handle Excel’s 1-based indexing
- Use modulo `26` to get the current character
- Convert the value to a character using ASCII mapping
- Build the result string from right to left

### Time Complexity
- **O(log₍26₎ n)**  
  - Number of digits in base-26 representation

### Space Complexity
- **O(1)**  
  - Excluding the output string


```python
class Solution:
    def convertToTitle(self, columnNumber: int) -> str:
        result = ""

        while columnNumber > 0:
            columnNumber -= 1              # adjust for 1-based indexing
            result = chr(columnNumber % 26 + ord('A')) + result
            columnNumber //= 26

        return result
```

```python
solution = Solution()
test_convert_to_title(solution)
```
