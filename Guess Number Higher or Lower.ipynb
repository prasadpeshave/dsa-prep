## Problem Statement

You are playing a guessing game. I pick a number from `1` to `n`, and you must guess it.

You can call the API `guess(num)` which returns:
- `-1` if `num` is **higher** than the picked number
- `1` if `num` is **lower** than the picked number
- `0` if `num` is **equal** to the picked number

Your task is to find the picked number using the minimum number of calls.

## Examples

### Example 1
**Input:** `n = 10, pick = 6`

**Output:** `6`

---

### Example 2
**Input:** `n = 1, pick = 1`

**Output:** `1`

---

### Example 3
**Input:** `n = 2, pick = 1`

**Output:** `1`

---

## Constraints

- `1 <= n <= 2³¹ - 1`
- `1 <= pick <= n`

# Mock Guess API (for local testing)
---

```python
PICK = 6

def guess(num: int) -> int:
    if num > PICK:
        return -1
    elif num < PICK:
        return 1
    else:
        return 0
```

# Approach
---

## Binary Search

### Strategy
- Search in the range `[1, n]`
- Pick the middle number
- Use the API result to eliminate half the range

### Why Binary Search?
- Each API call tells us whether to go left or right
- The search space is sorted
- This guarantees the minimum number of guesses

### Time Complexity
- **O(log n)**

### Space Complexity
- **O(1)**

```python
class Solution:
    def guessNumber(self, n: int) -> int:
        left, right = 1, n

        while left <= right:
            mid = (left + right) // 2
            res = guess(mid)

            if res == 0:
                return mid
            elif res == -1:
                right = mid - 1
            else:
                left = mid + 1
```

```python
solution = Solution()
print(solution.guessNumber(10))
```
