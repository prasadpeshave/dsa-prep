## Problem Statement

You are given an integer array `nums` where the largest integer is unique.

Determine whether the largest element in the array is **at least twice as much as every other number**.

- If it is, return the **index of the largest element**
- Otherwise, return `-1`

## Examples

### Example 1
**Input:**  
`nums = [3,6,1,0]`

**Output:**  
`1`

---

### Example 2
**Input:**  
`nums = [1,2,3,4]`

**Output:**  
`-1`

---

## Constraints

- `2 <= nums.length <= 50`
- `0 <= nums[i] <= 100`
- The largest element in nums is unique

# Test Cases
---

```python
def test_dominant_index(solution):
    test_cases = [
        ([3,6,1,0], 1),
        ([1,2,3,4], -1),
        ([0,0,0,1], 3),
        ([1,0], 0),
    ]

    for nums, expected in test_cases:
        result = solution.dominantIndex(nums)
        assert result == expected, f"Failed for {nums}: expected {expected}, got {result}"

    print("All test cases passed!")
```

# Approach
---

## Sort and Compare

### Strategy
- Sort the array in descending order
- Compare the largest value with twice the second-largest value
- If condition holds, return the index of the largest element

### Why This Works
- The largest element is guaranteed to be unique
- Only the second-largest value matters for the comparison

### Time Complexity
- **O(n log n)** due to sorting

### Space Complexity
- **O(n)** for the sorted array

```python
from typing import List

class Solution:
    def dominantIndex(self, nums: List[int]) -> int:
        sorted_nums = sorted(nums, reverse=True)
        if sorted_nums[0] >= 2 * sorted_nums[1]:
            return nums.index(sorted_nums[0])
        return -1
```

```python
solution = Solution()
test_dominant_index(solution)
```
