# Reverse Integer

---

## Problem Statement

Given a signed 32-bit integer `x`, return `x` with its digits reversed.

If reversing `x` causes the value to go outside the signed 32-bit integer range `[-2³¹, 2³¹ - 1]`, then return `0`.

Assume the environment does not allow you to store 64-bit integers.

## Examples

### Example 1
**Input:**  
`x = 123`

**Output:**  
`321`

---

### Example 2
**Input:**  
`x = -123`

**Output:**  
`-321`

---

### Example 3
**Input:**  
`x = 120`

**Output:**  
`21`

---

## Constraints

- `-2³¹ <= x <= 2³¹ - 1`

# Test Cases
---

```python
def test_reverse_integer(solution):
    test_cases = [
        (123, 321),
        (-123, -321),
        (120, 21),
        (0, 0),
        (1534236469, 0),  # overflow case
    ]

    for x, expected in test_cases:
        result = solution.reverse(x)
        assert result == expected, f"Failed for input {x}: expected {expected}, got {result}"

    print("All test cases passed!")
```

# Brute Force
---

## Brute Force Approach

### Strategy
- Extract the sign of the number
- Reverse the digits of the absolute value using string manipulation
- Reapply the original sign
- Validate the result against 32-bit integer bounds

### Time Complexity
- **O(n)**  
  - `n` is the number of digits in the integer

### Space Complexity
- **O(n)**  
  - Due to string conversion for digit reversal


```python
class BruteForceSolution:
    def reverse(self, x: int) -> int:
        # Determine the sign of the number
        sign = -1 if x < 0 else 1

        # Work with absolute value for digit reversal
        x = abs(x)

        # Reverse the digits using string manipulation
        reversed_val = int(str(x)[::-1]) * sign

        # Check for 32-bit signed integer overflow
        if not (-2**31 <= reversed_val <= 2**31 - 1):
            return 0

        return reversed_val
```

```python
solution = BruteForceSolution()
test_reverse_integer(solution)
```
