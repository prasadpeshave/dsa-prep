## Problem Statement

Given an integer array `nums`, return the **third distinct maximum number** in the array.

- If the third distinct maximum does not exist, return the **maximum number**.

## Examples

### Example 1
**Input:**  
`nums = [3,2,1]`

**Output:**  
`1`

---

### Example 2
**Input:**  
`nums = [1,2]`

**Output:**  
`2`

---

### Example 3
**Input:**  
`nums = [2,2,3,1]`

**Output:**  
`1`

---

## Constraints

- `1 <= nums.length <= 10⁴`
- `-2³¹ <= nums[i] <= 2³¹ - 1`

# Test Cases
---

```python
def test_third_max(solution):
    test_cases = [
        ([3,2,1], 1),
        ([1,2], 2),
        ([2,2,3,1], 1),
        ([1,1,1], 1),
        ([5,2,2], 5),
    ]

    for nums, expected in test_cases:
        result = solution.thirdMax(nums)
        assert result == expected, f"Failed for {nums}: expected {expected}, got {result}"

    print("All test cases passed!")
```

# Approach
---

## Using a Set and Sorting

### Strategy
- Remove duplicates using a set
- Sort the distinct values in descending order
- If there are at least three distinct numbers, return the third one
- Otherwise, return the maximum

### Time Complexity
- **O(n log n)** due to sorting

### Space Complexity
- **O(n)** for storing distinct values

```python
from typing import List

class Solution:
    def thirdMax(self, nums: List[int]) -> int:
        distinct = list(set(nums))
        distinct.sort(reverse=True)

        if len(distinct) >= 3:
            return distinct[2]
        return distinct[0]
```

```python
solution = Solution()
test_third_max(solution)
```
