## Problem Statement

An **ugly number** is a positive integer which does not have a prime factor other than `2`, `3`, and `5`.

Given an integer `n`, return `true` if `n` is an ugly number.

## Examples

### Example 1
**Input:**  
`n = 6`

**Output:**  
`true`

**Explanation:**  
`6 = 2 × 3`

---

### Example 2
**Input:**  
`n = 1`

**Output:**  
`true`

**Explanation:**  
`1` has no prime factors.

---

### Example 3
**Input:**  
`n = 14`

**Output:**  
`false`

**Explanation:**  
`14 = 2 × 7`, includes prime factor `7`.

---

## Constraints

- `-2^31 <= n <= 2^31 - 1`

# Test Cases
---

```python
def test_is_ugly(solution):
    test_cases = [
        (6, True),
        (1, True),
        (14, False),
        (8, True),
        (0, False),
        (-6, False),
    ]

    for n, expected in test_cases:
        result = solution.isUgly(n)
        assert result == expected, f"Failed for n={n}: expected {expected}, got {result}"

    print("All test cases passed!")
```

# Corrected Approach
---

## Divide by Allowed Prime Factors

### Strategy
- If `n` is less than or equal to `0`, return `false`
- Repeatedly divide `n` by `2`, `3`, and `5` while divisible
- If the final value becomes `1`, it is an ugly number

### Why This Works
- Ugly numbers contain no prime factors other than `2`, `3`, and `5`
- Removing all occurrences of these factors should reduce the number to `1`

### Time Complexity
- **O(log n)**

### Space Complexity
- **O(1)**

```python
class Solution:
    def isUgly(self, n: int) -> bool:
        if n <= 0:
            return False

        for factor in [2, 3, 5]:
            while n % factor == 0:
                n //= factor

        return n == 1
```

```python
solution = Solution()
test_is_ugly(solution)
```
