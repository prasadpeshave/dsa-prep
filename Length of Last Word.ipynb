## Problem Statement

Given a string `s` consisting of words and spaces, return the length of the last word in the string.

A word is a maximal substring consisting of non-space characters only.

## Examples

### Example 1
**Input:**  
`s = "Hello World"`

**Output:**  
`5`

**Explanation:**  
The last word is `"World"` with length `5`.

---

### Example 2
**Input:**  
`s = "   fly me   to   the moon  "`

**Output:**  
`4`

**Explanation:**  
The last word is `"moon"` with length `4`.

---

### Example 3
**Input:**  
`s = "luffy is still joyboy"`

**Output:**  
`6`

**Explanation:**  
The last word is `"joyboy"` with length `6`.

---

## Constraints

- `1 <= s.length <= 10â´`
- `s` consists of only English letters and spaces `' '`.
- There will be at least one word in `s`.

# Test Cases
---

```python
def test_length_of_last_word(solution):
    test_cases = [
        ("Hello World", 5),
        ("   fly me   to   the moon  ", 4),
        ("luffy is still joyboy", 6),
        ("a", 1),
        ("a ", 1),
    ]

    for s, expected in test_cases:
        result = solution.lengthOfLastWord(s)
        assert result == expected, f"Failed for input '{s}': expected {expected}, got {result}"

    print("All test cases passed!")
```

# Approach
---

### Strategy
- Split the string by spaces
- Traverse the resulting list in reverse order
- Ignore empty strings caused by extra spaces
- Return the length of the first valid word encountered

### Time Complexity
- **O(n)**  
  - `n` is the length of the string

### Space Complexity
- **O(n)**  
  - Due to splitting the string into a list


```python
class BruteForceSolution:
    def lengthOfLastWord(self, s: str) -> int:
        for word in reversed(s.split(" ")):
            if word:
                return len(word)
```

```python
solution = BruteForceSolution()
test_length_of_last_word(solution)
```

```python
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        return len(s.split()[-1])
```

```python
solution = Solution()
test_length_of_last_word(solution)
```
